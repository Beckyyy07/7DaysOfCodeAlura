# Dia 2 - Limpeza de dados

## Criando uma nova coluna com os valores da localização, para refletir a respectiva classe geral na CDU.


000 a 099: Generalidades. Ciência e conhecimento.

100 a 199: Filosofia e psicologia.

200 a 299: Religião.

300 a 399: Ciências sociais.

400 a 499: Classe vaga. Provisoriamente não ocupada.

500 a 599: Matemática e ciências naturais.

600 a 699: Ciências aplicadas.

700 a 799: Belas artes.

800 a 899: Linguagem. Língua. Linguística.

900 a 999: Geografia. Biografia. História."

```python
CDU_lista = []
for CDU in emprestimos_completo['localizacao']:
  if(CDU < 100):
    CDU_lista.append('Generalidades')
  elif(CDU < 200):
    CDU_lista.append('Filosofia e psicologia')
  elif(CDU < 300):
    CDU_lista.append('Religião')
  elif(CDU < 400):
    CDU_lista.append('Ciências sociais')
  elif(CDU < 500):
    CDU_lista.append('Classe vaga')
  elif(CDU < 600):
    CDU_lista.append('Matemática e ciências naturais')
  elif(CDU < 700):
    CDU_lista.append('Ciências aplicadas')
  elif(CDU < 800):
    CDU_lista.append('Belas artes')
  elif(CDU < 900):
    CDU_lista.append('Linguagem')
  else:
    CDU_lista.append('Geografia. Biografia. História.')

emprestimos_completo['CDU'] = CDU_lista
emprestimos_completo.head(5)
```

# Dia 3 - Análise exploratória de dados e DateTime

```python
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
```

## Verificando relação entre quantidade de emprestimos e quantidade de exemplares emprestados

```python
emprestimos_completo['id_emprestimo'].value_counts()
```

## Quantidade de emprestimos

```python
qntd_emprestimos = len(emprestimos_completo['id_emprestimo'].drop_duplicates())
qntd_emprestimos
```

## Quantidade de exemplares emprestados

```python
qntd_emprestimos_exemplares = len(emprestimos_completo)
qntd_emprestimos_exemplares
```

## Criando Dataframe de datas

```python
emprestimos_data = pd.DataFrame(emprestimos_completo['data_emprestimo'].value_counts()).reset_index()

#Renomeando as colunas
emprestimos_data.columns = ['Data','Quantidade']

#modificando o tipo da coluna data_emprestimo para dateTime
emprestimos_data['Data'] = pd.to_datetime(emprestimos_data['Data'])

emprestimos_data
```

## Quantidade de exemplares emprestados por ano

```python
emprestimos_por_ano = emprestimos_data['Data'].dt.year
emprestimos_por_ano = emprestimos_por_ano.value_counts().groupby(['Data']).sum().sort_values(ascending = False).to_frame('Quantidade')
emprestimos_por_ano.index.name = 'Ano'
emprestimos_por_ano
```

```python
ax = sns.lineplot(data = emprestimos_por_ano , x='Ano',y='Quantidade')
ax.set(xlabel= 'Ano',ylabel='Quantidade')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))

ax.set_title('Quantidade de exemplares emprestados do SISBI por ano'+'\n',size=11,loc='center',weight='bold')

ax=ax
```

## Análise da visualização de emprestimos por ano

## Quantidade de emprestimos por mês

```python
emprestimos_por_mes = emprestimos_data['Data'].dt.month
emprestimos_por_mes = emprestimos_por_mes.value_counts().groupby(['Data']).sum().sort_index().to_frame('Quantidade')
emprestimos_por_mes.index.name = 'Mês'
emprestimos_por_mes
```

## Modificando os números por nomes dos meses

```python
# Dicionário para mapear números de meses para nomes abreviados
dicionario_meses = {1: 'Jan', 2: 'Fev', 3: 'Mar', 4: 'Abr',
                    5: 'Mai', 6: 'Jun', 7: 'Jul', 8: 'Ago',
                    9: 'Set', 10: 'Out', 11: 'Nov', 12: 'Dez'}
```

```python
# Mapear índices do DataFrame para nomes abreviados dos meses
emprestimos_por_mes.index = emprestimos_por_mes.index.map(dicionario_meses)
```

```python
emprestimos_por_mes
```

```python
ax = sns.lineplot(data=emprestimos_por_mes,x='Mês',y='Quantidade')
ax.set(xlabel=None,ylabel=None)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))

ax.set_title("Quantidade de exemplares emprestados do SISBI por mês"+"\n",size=11,loc='left',weight='bold')
ax.text(s='Período de 2010 a 2020',x=-0.5,y=265000,fontsize=9, ha='left',color='gray')

ax=ax
```

Análise da visualização de quantidade de exemplares emprestados do SISBI por mês

descubra quais meses são os que possuem a maior quantidade de empréstimos realizados.

## Qual é a quantidade de empréstimos por hora ao longo do dia ?

```python
emprestimos_por_hora = emprestimos_data['Data'].dt.hour
emprestimos_por_hora = emprestimos_por_hora.value_counts().groupby(['Data']).sum().sort_index().to_frame('Quantidade')
emprestimos_por_hora.index.name = 'Hora'
emprestimos_por_hora
```

```python
plt.figure(figsize=(15, 4))

ax = sns.barplot(data=emprestimos_por_hora,y='Quantidade',x='Hora',
                 palette = 'crest' ,hue='Quantidade',dodge=False)                  #Ordenar pela quantidade de exemplares emprestados
plt.legend([],[], frameon=False)                                                #Excluir a legenda do gráfico


ax.set(xlabel='Horário',ylabel=None)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
ax.set_title("Quantidade de exemplares emprestados do SISBI por faixa horária"+"\n",size=11,loc='left',weight='bold')
ax.text(s='Período de 2010 a 2020',x=-0.5,y=225000,fontsize=9, ha='left',color='gray')
ax=ax
```
