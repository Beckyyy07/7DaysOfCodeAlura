e com o percentual para cada variável.

```python

def tabela_frequencia(variavel):
  df_frequencia = pd.DataFrame(df_emprestimo_completo[variavel].value_counts())
  df_frequencia.columns = ['Quantidade']
  df_frequencia['Percentual'] = round((df_frequencia.Quantidade / df_frequencia.Quantidade.sum())*100,1) #tira percentual e define a quantidade de digitos depois da virgula

  return df_frequencia
'''
  Esta função irá gerar uma tabela de frequência com percentuais de acordo
  com a variável passada.

  variavel = variável categórica escolhida de dentro do conjunto de dados
  emprestimos_completo
'''
```

## Função para gerar box plot

```python
def gera_boxplot(dataset, x, y, titulo):
  '''
  Dataset = conjunto de dados do gráfico
  x = valor do eixo x do gráfico
  y = valor do eixo y do gráfico
  título = título do gráfico
  subtitulo = subtitulo do texto
  '''

  sns.set_theme(style="whitegrid",font_scale=1.3, palette = 'icefire')
  plt.figure(figsize=(20,6))

  grafico = sns.boxplot(x=x, y=y, data=dataset, linecolor = 'Black')
  grafico.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
  sns.despine()
  grafico.set_title(titulo, fontsize=18)

  plt.show(grafico)
```
