# Dia 2 - Limpeza de dados

## Criando coluna com os valores da localização, refletindo a respectiva classe na CDU.

```python
CDU_list = []
for CDU in df_emprestimo_completo['localizacao']:
  if (CDU < 100):
    CDU_list.append('Generalidades')
  elif (CDU < 200):
    CDU_list.append('Filosofia e psicologia')
  elif (CDU < 300):
    CDU_list.append('Religião')
  elif (CDU < 400):
    CDU_list.append('Ciências sociais')
  elif (CDU < 500):
    CDU_list.append('Classe vaga')
  elif (CDU < 600):
    CDU_list.append('Matemática e ciências naturais')
  elif (CDU < 700):
    CDU_list.append('Ciências aplicadas')
  elif (CDU < 800):
    CDU_list.append('Belas artes')
  elif (CDU < 900):
    CDU_list.append('Linguagem')
  else:
    CDU_list.append('Geografia')

df_emprestimo_completo['CDU'] = CDU_list
```

## Quantidade de emprestimos

```python
#Criando DataFrame apenas com as colunas data e quantidade de emprestimos por data
data_emprestimo = pd.DataFrame(df_emprestimo_completo['data_emprestimo'].value_counts()).reset_index()
#renomeando colunas
data_emprestimo.columns = ['Data','Quantidade']
#transformando coluna em datetime
data_emprestimo['Data'] = pd.to_datetime(data_emprestimo['Data'])
data_emprestimo
```

### Quantidade de exemplares emprestados por ano

```python
emprestimos_por_ano = data_emprestimo.groupby(data_emprestimo['Data'].dt.year)['Quantidade'].sum().reset_index()
emprestimos_por_ano.columns = ['Ano', 'Quantidade']
emprestimos_por_ano
```

### Gráfico anual

```python
grafico_ano = sns.lineplot(data = emprestimos_por_ano, x = 'Ano', y = 'Quantidade')
grafico_ano.set_title('Quantidade de emprestimos por ano')
grafico_ano.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
plt.show(grafico_ano)
```

### Análise anual

De 2010 a 2013, os empréstimos aumentaram consistentemente, atingindo o ponto mais alto em 2013.

A partir desse ano, começaram a declinar continuamente. Em 2017, houve um breve aumento, mas nos anos seguintes, os empréstimos voltaram a diminuir. Em 2019, essa queda se intensificou.

É importante ressaltar que a análise de 2020 é limitada devido à falta de dados suficientes disponíveis para esse ano.

### Quantidade de exemplares emprestados por mês

```python
emprestimos_por_mes = data_emprestimo.groupby(data_emprestimo['Data'].dt.month)['Quantidade'].sum().to_frame()
emprestimos_por_mes.index.name = 'Mes'
emprestimos_por_mes
```

```python
#Modificando os números por nomes dos meses
dicionario_meses = {1:'Jan',2:'Fev',3:'Mar',4:'Abr',5:'Mai', 6:'Jun', 7:'Jul',8:'Ago', 9:'Set',10:'Out',11:'Nov',12:'Dez'}
#Mapeando e atribuindo novos rótulos aos índices
emprestimos_por_mes.index = emprestimos_por_mes.index.map(dicionario_meses)
emprestimos_por_mes
```

### Gráfico Mensal

```python
grafico_mes = sns.lineplot(data = emprestimos_por_mes, x = 'Mes', y = 'Quantidade')
grafico_mes.set_title('Quantidade de emprestimos por mês')
grafico_mes.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
plt.show(grafico_mes)
```

### Análise mensal

Os meses com maior atividade de empréstimos foram **MARÇO** e **AGOSTO**.

Janeiro, julho e dezembro são meses de férias, resultando em baixos volumes de empréstimos.

Fevereiro e março registraram um aumento nas operações de empréstimo, sendo março o mais movimentado.

No entanto, a partir de março, os empréstimos começaram a diminuir. Em agosto, houve um pico de atividade de empréstimos, seguido por uma queda subsequente.

Considerando a baixa demanda durante **JANEIRO** e **DEZEMBRO**, esses meses poderiam ser ideais para programar **FÉRIAS** dos funcionários.

Os outros meses de menor atividade de empréstimos, como maio, junho, outubro e novembro, poderiam ser aproveitados para realizar **OUTRAS ATIVIDADES**.

### Quantidade de exemplares emprestados por hora

```python
emprestimos_por_hora = data_emprestimo.groupby(data_emprestimo['Data'].dt.hour)['Quantidade'].sum().to_frame()
emprestimos_por_hora.index.name = 'Horas'
emprestimos_por_hora
```

### Gráfico de horas

```python
grafico_horas = sns.barplot(data = emprestimos_por_hora, x = 'Horas', y = 'Quantidade',  palette = 'rocket_r', hue = 'Quantidade')
grafico_horas.set_title('Quantidade de emprestimos por hora')
grafico_horas.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
plt.show(grafico_horas)
```

### Análise por horário

Com base na análise do gráfico, podemos observar que os horários de pico de movimento são às **10:00** e às **16:00**.

Para otimizar a realização de outras atividades, os melhores horários seriam no início da manhã, entre 7 e 9 horas, e no início da noite, entre 19 e 21 horas. Esses períodos apresentam menor atividade de empréstimos, permitindo maior disponibilidade para outras tarefas.
